```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication

# Set up SMTP server credentials
smtp_server = 'smtp.gmail.com'
smtp_port = 587
email = 'syed.suhana@btech.christuniversity.in'
password = '72156328'

# Set up email message
recipient = 'suhanasyed2003@gmail.com'
subject = 'Yolov8 model Prediction'
body = 'This email contains image of various roads containing emergency vehicles!!!'

# Create a multipart message container and add the message and attachment to it
message = MIMEMultipart()
message['From'] = email
message['To'] = recipient
message['Subject'] = subject
message.attach(MIMEText(body))
with open('/content/output.jpeg', 'rb') as attachment:
    file = MIMEApplication(attachment.read(), _subtype='jpeg')
    file.add_header('Content-Disposition', 'attachment', filename='output.jpeg')
    message.attach(file)

# Log in to SMTP server and send message
with smtplib.SMTP(smtp_server, smtp_port) as server:
    server.starttls()
    server.login(email, password)
    server.sendmail(email, recipient, message.as_string())
    print('Email sent successfully with attachment!')
```
